```python
from langsmith import Client
```

### Attached feedback to a specific trace in LangSmith

```python
TRACE_ID = "019dc900-9a98-7921-86a3-f1433c16c455"
```

```python
langsmith = Client()
```

```python
langsmith.create_feedback(
    run_id = TRACE_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
langsmith.create_feedback(
    run_id = TRACE_ID,
    key="comment",
    value="This is a long form free text feedback.",
    feedback_source_type="api"
)
```

```python
langsmith.create_feedback(
    run_id = TRACE_ID,
    key="comment_with_score",
    value="This is a long form free text feedback.",
    score=1,
    feedback_source_type="api"
)
```
